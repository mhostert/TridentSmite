```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
stuff1 = np.genfromtxt("fluxes/T2Kflux2013/t2kflux_2013_nd280_horn205kA.txt",unpack=True,skip_header=3)
stuff2 = np.genfromtxt("fluxes/T2Kflux2013/t2kflux_2013_nd280_horn250kA.txt",unpack=True,skip_header=3)
```

```python
_, e1, e2,numu, numubar, nue , nuebar = stuff1 
_, e1_2, e2_2,numu_2, numubar_2, nue_2 , nuebar_2 = stuff2 
```

```python
plt.plot(e1+(e2-e1)/2.0,numu,color='indigo')
plt.plot(e1+(e2-e1)/2.0,numubar,color='green')

plt.plot(e1_2+(e2_2-e1_2)/2.0,numu_2,color='indigo',ls='--')
plt.plot(e1_2+(e2_2-e1_2)/2.0,numubar_2,color='green',ls='--')

plt.xlim(0,3)
# plt.yscale('log')
# plt.xscale('log')
```

```python
np.savetxt("fluxes/ND280_horn205kA.dat",np.array([e1+(e2-e1)/2.0,nue,numu,numu*0,nuebar,numubar,numu*0]).T)
np.savetxt("fluxes/ND280_horn2505kA.dat",np.array([e1+(e2-e1)/2.0,nue,numu,numu*0,nuebar,numubar,numu*0]).T)
```
